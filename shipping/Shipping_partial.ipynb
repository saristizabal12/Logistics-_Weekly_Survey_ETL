```python
import pandas as pd

df = pd.read_excel(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/shipping_partial_output.xlsx')

row_names = ['FTZ', 'Main', 'Mercado Libre', 'Ricardo', 'Joseph', 'Wilenis','Alan','FTZ A1','FTZ A2']
column_to_check = 'SO #'

# Ensure 'Location' column exists and is of object dtype
if 'Location' not in df.columns:
    df['Location'] = ''
else:
    df['Location'] = df['Location'].astype('object')

current_location = None

for i in range(len(df)):
    value = df.at[i, column_to_check]

    if pd.notna(value) and str(value).strip() in row_names:
        current_location = str(value).strip()
        df.at[i, 'Location'] = current_location
        continue

    if current_location and pd.notna(value):
        df.at[i, 'Location'] = current_location
    elif pd.isna(value):
        current_location = None

df.to_excel(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/shipping_partial_output.xlsx', index=False)
```
