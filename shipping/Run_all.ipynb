```python
import pandas as pd

path1 = r"C:/Users/saristizabal/Documents/Reports/Gitfolder/Logistics-_Weekly_Survey_ETL/Surveys/testing/output_yes/shipping_output.xlsx"
path2 = r"C:/Users/saristizabal/Documents/Reports/Gitfolder/Logistics-_Weekly_Survey_ETL/Surveys/testing/output_yes/shipping_entered_output.xlsx"
path3 = r"C:/Users/saristizabal/Documents/Reports/Gitfolder/Logistics-_Weekly_Survey_ETL/Surveys/testing/output_yes/shipping_issued_output.xlsx"

# Read files
df1 = pd.read_excel(path1, engine="openpyxl")
df2 = pd.read_excel(path2, engine="openpyxl")
df3 = pd.read_excel(path3, engine="openpyxl")

# Clean headers (important)
for df in (df1, df2, df3):
    df.columns = df.columns.astype(str).str.strip().str.replace("\ufeff", "", regex=False)

# Combine
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Save output
combined_df.to_excel(
    r"C:/Users/saristizabal/Documents/Reports/Gitfolder/Logistics-_Weekly_Survey_ETL/Surveys/testing/output_yes/shipping_output.xlsx",
    index=False
)
```
