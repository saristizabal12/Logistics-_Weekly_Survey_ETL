```python
import pandas as pd

shipping_path = r"C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Sell_Thru_shipping.csv"
sellthru_in = r"C:/Users/saristizabal/Documents/Reports/Sell Thru/testing/input_yes/Sell_Thru.xlsx"
sellthru_out = r"C:/Users/saristizabal/Documents/Reports/Sell Thru/testing/output_yes/Sell_Thru_shipping.xlsx"

ship = pd.read_csv(shipping_path)
ship.columns = ship.columns.astype(str).str.strip().str.replace("\ufeff", "", regex=False)

def pick_col(df, candidates):
    lower_map = {c.strip().lower(): c for c in df.columns}
    for name in candidates:
        key = name.strip().lower()
        if key in lower_map:
            return lower_map[key]
    return None

model_col = pick_col(ship, ["Model", "SO #", "SO#", "Product", "Item", "Model Friendly Name"])
qty_col = pick_col(ship, ["QTY", "Quantity", "Qty", "QUANTITY"])

if model_col is None:
    model_col = ship.columns[0]

if qty_col is None:
    raise ValueError(f"Could not find a quantity column. Columns found: {ship.columns.tolist()}")

ship[model_col] = ship[model_col].astype(str).str.strip()
ship[qty_col] = pd.to_numeric(ship[qty_col], errors="coerce").fillna(0)

ship_summary = ship.groupby(model_col, as_index=False)[qty_col].sum()
ship_summary = ship_summary.rename(columns={model_col: "Model", qty_col: "QTY"})

sell_thru = pd.read_excel(sellthru_in, engine="openpyxl")
sell_thru.columns = sell_thru.columns.astype(str).str.strip().str.replace("\ufeff", "", regex=False)

sell_thru["Model"] = sell_thru["Model"].astype(str).str.strip()
sell_thru["On Order"] = pd.to_numeric(sell_thru["On Order"], errors="coerce").fillna(0)

sell_thru = sell_thru.merge(ship_summary, on="Model", how="left")

sell_thru["On Order"] = sell_thru["On Order"] + sell_thru["QTY"].fillna(0)

sell_thru.drop(columns=["QTY"], inplace=True)

sell_thru.to_excel(sellthru_out, index=False)
```
