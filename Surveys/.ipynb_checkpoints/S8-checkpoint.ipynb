```python
import pandas as pd

# Step 1: Load CSV with no header
df_raw = pd.read_csv(
    r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S8.csv',
    header=None
)

# Step 2: Inspect first 10 rows to find the correct header row
print(df_raw.head(10))
```

```python
# verifying path
df = pd.read_csv(
    r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S8.csv',
    skiprows=1
)

#Dropping completely unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Reset index
df.reset_index(drop=True, inplace=True)

df = df.iloc[:,2:]

# df = df.drop(df.index[1]).reset_index(drop=True)
df = df.drop(df.index[0]).reset_index(drop=True)

print(df.head())
```

```python
# Save cleaned DataFrame to a new CSV file
df.to_csv(
    r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S8.csv',
    index=False
)
```

```python
# Ensure the first column is treated as string
first_col = df.columns[0]

df = df[~df[first_col].astype(str).str.startswith('AC SM')].reset_index(drop=True)
df.to_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S8.csv', index=False)
```

```python
first_col = df.columns[0]

df = df[~df[first_col].astype(str).str.startswith('Service Fee')].reset_index(drop=True)

df.to_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S8.csv', index=False)
```
