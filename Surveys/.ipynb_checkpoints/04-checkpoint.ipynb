```python
# Load the file
df = pd.read_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/4.csv')

# Group by 'MODEL' and count occurrences
model_totals = df['MODEL'].value_counts().reset_index()
model_totals.columns = ['MODEL', 'AMOUNT']

# Sort in ascending order
model_totals = model_totals.sort_values(by='AMOUNT', ascending=True).reset_index(drop=True)

# Save the cleaned result
model_totals.to_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Place_holder/4_clean.csv', index=False)
```

```python
# Step 1: Reload and sort the IMEI data
model_totals = pd.read_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Place_holder/4_clean.csv')
model_totals = model_totals.sort_values(by='MODEL', ascending=True).reset_index(drop=True)

# (Optional) Save the sorted IMEI file again
model_totals.to_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Place_holder/4_clean.csv', index=False)

# Step 2: Load the sorted IMEI file and the Sales file
imei_df = pd.read_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Place_holder/4_clean.csv')
sales_df = pd.read_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S4.csv')

# Step 3: Merge IMEI and Sales data based on MODEL and PRODUCT
merged_df = pd.merge(sales_df, imei_df, left_on='Product', right_on='MODEL', how='inner')

# Step 4: Update 'Quantity' with 'AMOUNT'
merged_df['Quantity'] = merged_df['AMOUNT']

# Step 5: Drop unnecessary columns
merged_df = merged_df.drop(columns=['MODEL', 'AMOUNT'])

# Step 6: Save updated sales file
merged_df.to_csv(r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/S4.csv', index=False)

print("All IMEI data sorted and Sales file updated with matching AMOUNT values.")
```

```python
import os
import pandas as pd

# File paths
name_map_csv = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/input_yes/name_map_csv/names.csv'
csv_folder = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/'

# Read the name mapping file (with header)
name_map = pd.read_csv(name_map_csv, header=0, usecols=[0, 1])
name_map.columns = ['OldName', 'NewName']

# Clean the values and handle missing data
name_map['OldName'] = name_map['OldName'].fillna('').astype(str).str.strip()
name_map['NewName'] = name_map['NewName'].fillna('').astype(str).str.strip()

# List actual files in the output folder
existing_files = os.listdir(csv_folder)

# Loop and rename matching files
for _, row in name_map.iterrows():
    old_filename = row['OldName']
    new_filename = row['NewName'] + '.csv'

    # Skip empty entries
    if not old_filename or not row['NewName']:
        print(f"Skipping row with missing name: {row}")
        continue

    old_path = os.path.join(csv_folder, old_filename)
    new_path = os.path.join(csv_folder, new_filename)

    if old_filename in existing_files:
        os.rename(old_path, new_path)
        print(f'Renamed: {old_filename} → {new_filename}')
    else:
        print(f'File not found: {old_filename} — skipped')
```
