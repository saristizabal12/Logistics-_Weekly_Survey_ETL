```python
import pandas as pd

# File paths
combined_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/combined.csv'
survey_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/input_yes/Sales_Team_Survey_Format.xlsx'
output_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Survey_testing.xlsx'

# Load files
combined_df = pd.read_csv(combined_file)
survey_df = pd.read_excel(survey_file)

# Clean column names
combined_df.columns = combined_df.columns.str.strip()
survey_df.columns = survey_df.columns.str.strip()

# Clean needed fields
combined_df['Name'] = combined_df['Name'].astype(str).str.strip()
combined_df['Model Friendly Name'] = combined_df['Model Friendly Name'].astype(str).str.strip()
combined_df['Quantity'] = pd.to_numeric(combined_df['Quantity'], errors='coerce').fillna(0)

survey_df['Name'] = survey_df['Name'].astype(str).str.strip()
survey_df['Model Friendly Name'] = survey_df['Model Friendly Name'].astype(str).str.strip()

# Make sure Quantity exists in survey file
if 'Quantity' not in survey_df.columns:
    survey_df['Quantity'] = 0

survey_df['Quantity'] = pd.to_numeric(survey_df['Quantity'], errors='coerce').fillna(0)

# Group combined.csv by BOTH Name and Model Friendly Name
grouped_df = (
    combined_df
    .groupby(['Name', 'Model Friendly Name'], as_index=False)['Quantity']
    .sum()
)

# Merge on BOTH columns
updated_df = survey_df.merge(
    grouped_df,
    on=['Name', 'Model Friendly Name'],
    how='left',
    suffixes=('', '_new')
)

# Only place quantity where both columns matched
updated_df['Quantity'] = updated_df['Quantity_new'].fillna(updated_df['Quantity'])

# Drop helper column
updated_df.drop(columns=['Quantity_new'], inplace=True)

# Save output
updated_df.to_excel(output_file, index=False)

print("Done. Quantities were updated only where both 'Name' and 'Model Friendly Name' matched.")
```

```python
import pandas as pd
import os

# File paths
combined_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/combined_shipping.xlsx'
survey_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Survey_testing.xlsx'
output_file = r'C:/Users/saristizabal/Documents/Reports/Surveys/testing/output_yes/Survey_testing_updated.xlsx'

# Check files exist
if not os.path.exists(combined_file):
    raise FileNotFoundError(f"Combined file not found: {combined_file}")

if not os.path.exists(survey_file):
    raise FileNotFoundError(f"Survey file not found: {survey_file}")

# Load files
combined_df = pd.read_excel(combined_file)
survey_df = pd.read_excel(survey_file)

# Clean column names
combined_df.columns = combined_df.columns.str.strip()
survey_df.columns = survey_df.columns.str.strip()

# Clean required columns in combined file
combined_df['Name'] = combined_df['Name'].astype(str).str.strip()
combined_df['Model Friendly Name'] = combined_df['Model Friendly Name'].astype(str).str.strip()
combined_df['Shipping'] = pd.to_numeric(combined_df['Shipping'], errors='coerce').fillna(0)

# Clean required columns in survey file
survey_df['Name'] = survey_df['Name'].astype(str).str.strip()
survey_df['Model Friendly Name'] = survey_df['Model Friendly Name'].astype(str).str.strip()

# Make sure Quantity column exists in survey
if 'Shipping' not in survey_df.columns:
    survey_df['Shipping'] = 0

survey_df['Shipping'] = pd.to_numeric(survey_df['Shipping'], errors='coerce').fillna(0)

# Group combined file by Name and Model Friendly Name
grouped_df = (
    combined_df
    .groupby(['Name', 'Model Friendly Name'], as_index=False)['Shipping']
    .sum()
)

# Merge only on matching Name and Model Friendly Name
updated_df = survey_df.merge(
    grouped_df,
    on=['Name', 'Model Friendly Name'],
    how='left',
    suffixes=('', '_new')
)

# Replace Quantity only where a match exists
updated_df['Shipping'] = updated_df['Shipping_new'].fillna(updated_df['Shipping'])

# Drop helper column
updated_df.drop(columns=['Shipping_new'], inplace=True)

# Save output
updated_df.to_excel(output_file, index=False)

print("Done. Quantities updated only where both 'Name' and 'Model Friendly Name' matched.")
print(f"Saved file: {output_file}")
```
